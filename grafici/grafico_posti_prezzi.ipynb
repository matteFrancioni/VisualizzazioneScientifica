```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv('cars.csv')

prices = db.groupby("Seats")["Cars Prices"].mean() / 10000

plt.barh(prices.index, prices.values, color = "#E8435B")
plt.yticks(np.arange(0, 13))
plt.xticks(np.arange(0, 81, 5))
plt.ylabel("Posti")
plt.xlabel("Prezzo x 10000")
plt.title("Posti/Prezzo")
plt.show()
```
