```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv('cars.csv')

fuels = db['Fuel Types'].unique()

summary = pd.DataFrame({
    fuel: [
        db[db['Fuel Types'] == fuel]['Total Speed'].max(),
        db[db['Fuel Types'] == fuel]['Total Speed'].mean(),
        db[db['Fuel Types'] == fuel]['Total Speed'].min()
    ]
    for fuel in fuels
}, index=["Max", "Mean", "Min"])

summary.T.plot(kind="bar")
plt.ylabel("Top Speed")
plt.title("Fuel vs Top speed")
plt.show()
```
