```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv('cars.csv')

engines_freq = db['Engines'].value_counts()
common_engines = engines_freq[engines_freq > 20].index

engines = db[db['Engines'].isin(common_engines)]
mean_speeds = engines.groupby("Engines")["Total Speed"].mean()
plt.barh(mean_speeds.index, mean_speeds.values, color= "#43E8A6")
plt.xlabel("Mean Speed")
plt.ylabel("Engine")
plt.title("Engine/ Mean Speed")
plt.show()
```
