```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv('cars.csv')

meno100k = db[db['Cars Prices'] < 250000]
plt.scatter(meno100k['Performance(0 - 100 )KM/H'], meno100k['Cars Prices'], alpha = 0.7, color = "#6CCAEB", edgecolors = "#43B6E8" )
plt.xlabel("0-100 km/h")
plt.ylabel("Prezzo")
plt.title("Prezzo vs Performance")
plt.show()
```
