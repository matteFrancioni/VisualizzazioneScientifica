```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv('cars.csv')

brands = ["FERRARI", "ROLLS ROYCE", "PORSCHE", "NISSAN"]

summary = pd.DataFrame({
    brand: [
        db[db['Company Names'] == brand]['Cars Prices'].max()/10000,
        db[db['Company Names'] == brand]['Cars Prices'].mean()/10000,
        db[db['Company Names'] == brand]['Cars Prices'].min()/10000
    ]
    for brand in brands
}, index=["Max", "Mean", "Min"])

summary.T.plot(kind="bar")
plt.ylabel("Prezzo X 10000")
plt.title("Prezzi Auto (Max / Mean / Min)")
plt.show()
```
