```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv('cars.csv')

brands = ["KIA", "VOLKSWAGEN", "CHEVROLET", "PEUGEOT"]

summary = pd.DataFrame({
    brand: [
        db[db['Company Names'] == brand]['Cars Prices'].max()/1000,
        db[db['Company Names'] == brand]['Cars Prices'].mean()/1000,
        db[db['Company Names'] == brand]['Cars Prices'].min()/1000
    ]
    for brand in brands
}, index=["Max", "Mean", "Min"])

summary.T.plot(kind="bar")
plt.ylabel("Prezzo X 1000")
plt.title("Prezzi Auto (Max / Mean / Min)")
plt.show()
```
